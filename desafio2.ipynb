```python
def agente_financeiro():
    print("💰 Bem-vindo ao seu Assistente Financeiro Inteligente\n")

    # Coleta de dados
    nome = input("Qual seu nome? ")
    saldo = float(input("Qual seu saldo atual? R$ "))
    gastos = float(input("Quanto você gasta por mês? R$ "))
    objetivo = input("Qual seu objetivo financeiro? (economizar/investir/quitar dívidas) ")

    # Organização dos dados
    usuario = {
        "nome": nome,
        "saldo": saldo,
        "gastos": gastos,
        "objetivo": objetivo
    }

    print(f"\nOlá, {usuario['nome']}! Vamos analisar sua situação financeira:\n")

    # Análise financeira
    if usuario["saldo"] < 0:
        print("⚠️ Situação crítica: saldo negativo.")
        print("Sugestão: priorize quitar dívidas e evite novos gastos.\n")

    elif usuario["saldo"] < usuario["gastos"]:
        print("⚠️ Atenção: seus gastos são maiores que seu saldo.")
        print("Sugestão: reduzir despesas ou buscar renda extra.\n")

    else:
        print("✅ Situação estável.")
        print("Sugestão: você pode começar a investir parte do seu dinheiro.\n")

    # Personalização baseada no objetivo
    if usuario["objetivo"] == "economizar":
        print("💡 Dica: crie um controle de gastos e defina metas mensais.")

    elif usuario["objetivo"] == "investir":
        print("💡 Dica: estude renda fixa e comece com baixo risco.")

    elif usuario["objetivo"] == "quitar dívidas":
        print("💡 Dica: priorize dívidas com juros altos primeiro.")

    # Antecipação de risco
    if usuario["gastos"] > usuario["saldo"] * 0.8:
        print("\n🔎 Alerta: você está gastando quase todo seu dinheiro mensalmente.")

    # Segurança (anti-alucinação)
    print("\n⚠️ Aviso: Este sistema fornece sugestões baseadas nos dados informados e não substitui um especialista financeiro.")

# Execução do programa
agente_financeiro()
```
